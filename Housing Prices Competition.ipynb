```python

from sklearn.preprocessing import LabelEncoder # 문자 값들을 숫자로 한방에 바꿔주는 것 
le = LabelEncoder()
# 22:12 - 모든 열을 추가로 넣어보자 
for i in train.columns[train.dtypes==object]:
    le.fit(list(train[i]) + list(test[i]))
    train[i] = le.transform(list(train[i]))
    test[i] = le.transform(list(test[i]))

# le.fit(train['BsmtFinType1']) # 문제 : test셋에서만 있는 것들은 적용 x,  해결 :  같이 적용하는 게 필요함
# le.fit(pd.concat([train['BsmtFinType1'], test['BsmtFinType1']]))  # 문제 : test셋에서만 있는 것들은 적용 x,  해결 :  같이 적용하는 게 필요함
# -> 결측치가 있으면 이거 실행 안됨
# le.fit(list(train['BsmtFinType1']) + list(test['BsmtFinType1']))
# # -> list로 하면 결측치를 하나의 문자로(none)으로 바꿈
# train['BsmtFinType1'] = le.transform(list(train['BsmtFinType1']))
# test['BsmtFinType1'] = le.transform(list(test['BsmtFinType1']))
# -> 문자형 컬럼(도움이 될만한 컬럼을 뽑아서)을 숫자로 바꿔 모델이 학습을 할 수 있게 판을 깔아줌 ************ 중요
#

# 22:06 컬럼을 추가하면 데이터셋이 복잡해짐. 모델은 방치하고 = 이에 문자혈 컬럼을 숫자로 바꿔도 점수가 안좋게 나옴
# 모델을 동시에 학습을 시켜줘서 점수를 좋게 

```

```python
train2 = train[train.columns[train.dtypes!=object]]
train2.shape
```

```python
test.dtypes!=object
```

```python
test2 = test[test.columns[test.dtypes!=object]]
test2.shape
```

```python
train3 = train2.drop(['SalePrice','Id'], axis=1) # train, test 열 갯수 차이남. saleprice 지우고 열 갯수 맞춰줌
test3 = test2.drop(['Id'], axis=1) # id는 별 의미 없어서 삭제 
```

```python
# 모델 다음에 실행 
# 결측치 채우기 
#train3.fillna(-1)
#test3.fillna(-1)

# 함수 재정의 필요 
train3=train3.fillna(-1)
test3=test3.fillna(-1)
```

```python
# # 랜덤포레스트를 쓰는게 일반적, 모델링은 한 곳에서 
# from sklearn.ensemble import RandomForestRegressor
# rf = RandomForestRegressor(random_state = 42, n_estimators = 100)  # 랜포 특징이 결과 값이 항상 변해서 고정으로 
# # 22:07 하이퍼파라미터 에스티메이터 추가 기본 나무 10개 -> 100개
# # 22:10 문자형 변수를 숫자로 바꿔 + 모델도 학습, 더 좋게 
# rf.fit(train3, train['SalePrice'])
# result = rf.predict(test3)




# Input contains NaN, infinity or a value too large for dtype('float32') -> 결측치 존재(해결법 = 0, 중앙값, 평균값, ***다른 데이터와 겹치지 않게 넣어주는 것 = -1)
# 트리모델 학습은 데이터를 세분화하면서 트리가 널 값을 인식할 수 있게함 
# 데이터 도메인 지식 같은 게 없을 땐 모델을 이용한 방법도 존재함  
# 트리 모델은 중요한 변수를 알아서 더 가중치를 주어서 학습시킴
```

```python
# 22:26 부스팅을 왜 써야하는 가 
# 트리 - 
# 1) 배깅(랜덤포레스트 - 하나의 나무를 만들고, 10000개 중 8000개, 60개 등을 랜덤하게 뽑고, 만들고 여러 결과물들을 종합해서 봄 
# / 단점 - 서로서로 돕지 않음, 1번 나무가 잘못분류한 걸 2번이 반영 못함 ) / 

# 2) 부스팅(1번나무에서 잘못된 분류를 2번 나무에 반영, 수정해 학습함 = 나무가 많이 만들어질수록 성능이 좋아짐)
# / 단점 - train 셋에만 학습이 잘 될 수 있기에 과대적합 문제가 발생할 수 있음 
# 이에 부스팅에서는 하이퍼파라미터 세팅이 필수적이면서 중요함!
# 현업에서는 랜포 정도만 해도 됨
```

```python
from lightgbm import LGBMRegressor
lgb = LGBMRegressor(num_leaves = 27, n_estimators = 80 ) # 22:42 num_leaves 추가 / 25는 단순하게 접근 덜 질문하는 거임 = 덜 학습하는 게 좋다 # 17하니깐 더 안좋아짐 # 27
# 22:47 n_estimators 추가 나무의 갯수를 100 -> 120 늘렸을 때(트레인셋을 더 많이 보면서), 
# 점수 잘 나오면 train과 test 데이터 분포가 비슷? 아니면 다른걸까? = 비슷 비슷해지니깐 점수가 train에서도 잘나오고 test에서도 잘나옴 
# 나무의 개수를 줄이고 모델점수가 더 잘나오면, train 셋을 덜 학습하는 게 중요하다 즉, train과 test 데이터가 다르구나 
# 다른 모델 (선형모델 거리기반 ) 사용 + 부스터 앙상블 
lgb.fit(train3, train['SalePrice'])
result = lgb.predict(test3)

# 22:34 부스팅이 항상 랜포보다 좋게 나오는 건 아님
# 랜포부터시작하면, 점수가 준수하게 잘나오고, 그리고 어떤 변수가 중요하고 유의미한지 파악하기 쉬움
# 과적합을 줄이기 위해 나무의 깊이(=질문을 얼마나 많이 할 것이냐 = 데이터셋이 복잡함, non leaves = 31이 기본값)
# 랜포는 80개를 다쓰고, lgbm은 31개만 썻는데 점수가 더 나음 = 이 데이터셋은 단순함 = lgbm은 과적합이 나오기 마련인데 이건 점수가 더 좋게 나오는 건 데이터셋이 단순해서임
# 트레인 테스트 셋 비율 -> 제출에 제한 x 

```
